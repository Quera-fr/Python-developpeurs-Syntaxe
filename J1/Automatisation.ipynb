```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time

def get_article(text_input:str='Python') -> dict:
    """
    Fonction qui retourne dans un json les articles du site lesechos.fr


    Parametres
    -----
    text_input : str | Objet de la recherche sur le site les echos.

    Return
    ------
    dict : Dictionnaire {titre, img_url}
    """

    # Ouverture du navigateur
    driver = webdriver.Chrome()

    # Chargement du site lesechos.fr
    driver.get('https://www.lesechos.fr/')

    # Accepter les cookies
    driver.find_element(By.CLASS_NAME, 'inv-cmp-button.inv-font-btn.inv-bold.inv-mb_16').click()

    # Champ de recherche
    driver.find_element(By.XPATH,'//*[@id="app"]/header/div/div[2]/div[1]/div[1]/button[2]').click()

    time.sleep(3)

    # Recherche
    driver.find_element(By.CLASS_NAME, 'text-field__input').send_keys(text_input + Keys.ENTER)

    time.sleep(3)

    # Liste des articles
    articles = driver.find_elements(By.TAG_NAME, 'article')

    i = 0
    dict_articles = {}
    # Affichage des titre et des images
    for article in articles:
        dict_articles[i] = {
            'tile' : article.find_element(By.TAG_NAME, 'h3').text,
            'img_url' : article.find_element(By.TAG_NAME, 'img').get_attribute('src')
        }
        i +=1

    # Fermeture du navigateur
    driver.close()

    return dict_articles
```

```python
dict_article = get_article()
```

```python
import json 


with open('data.json', "w") as f:
    json.dump(dict_article, f)
```
